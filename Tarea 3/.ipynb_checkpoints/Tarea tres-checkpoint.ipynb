```python
#Instrucción 1
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('hipparcos-voidmain.csv')
data.head()
```

```python
#Instrucción 2

últimas=data.iloc[[-5,-4,-3,-2,-1]]
últimas
```

```python
#Instrucción 3

primeras=data.loc[:, ["Catalog",'HIP','Proxy']]
primeras
```

```python
#Instrucción 4

nueva_data= data.loc[:, ['HIP','RAhms', 'DEdms', 'Vmag', 'Plx','SpType', 'B-V']]
nueva_data.dropna(inplace=True)
nueva_data
```

```python
#Instrucción 5

nueva_data= nueva_data.rename(columns={'B-V' : 'BV'})
nueva_data
```

```python
#Instrucción 6

x=nueva_data["Vmag"].max()
print(x)
y=nueva_data["Vmag"].min()
y
```

```python
#Instrucción 7
nueva_data["Vmag"].sort_values()
```

```python
#Instrucción 8
estrellas= []
for i in nueva_data["Vmag"]:
    if i>10:
        estrellas.append(i)
#print(estrellas)
archivo= open("estrellas_filtro.csv", "w")
archivo.write(str(estrellas))
archivo.close()
```

```python
#Instrucción 9
estrellas_BV= nueva_data[(nueva_data.BV>0.3) & (nueva_data.BV<0.6)]
estrellas_BV.BV.count()
```

```python

fig, ax = plt.subplots(figsize=(10,12))

ax.set_xlabel('Índice BV')
ax.set_ylabel('Frecuencia')
plt.hist(estrellas_BV["BV"], bins=50)
```

```python
#Instrucción 10

nueva_data.insert(loc=4, column='Distancia', value='')
nueva_data
```

```python
nueva_data['Distancia'] = (1/(nueva_data['Plx']*10**-3))
import math
import numpy
nueva_data['AbsMag'] = (nueva_data.Vmag - 5 * numpy.log10(nueva_data.Distancia) + 5)
nueva_data.head()
```

```python
fig, ax = plt.subplots(figsize=(10,12))

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(15, -10)
ax.grid()
ax.set_title('H-R Diagram - Hipparcos Catalog')

ax.title.set_fontsize(20)
ax.set_xlabel('Índice de color BV')
ax.set_ylabel('Magnitud absoluta')
k= plt.scatter(nueva_data['BV'],nueva_data['AbsMag'], s=1)
#Instrucción 11
plt.savefig('H-R Diagram - Hipparcos Catalog')
```
